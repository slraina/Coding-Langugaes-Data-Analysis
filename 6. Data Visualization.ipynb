## Demo: How to work with database


Download database file.


```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list ALL TABLES 


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name 
FROM sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as Count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here

import seaborn as sns

QUERY = """
SELECT *
FROM master
"""

df = pd.read_sql_query(QUERY,conn)
df.head(3)
```

```python
# the read_sql_query runs the sql query and returns the data as a dataframe

plt.figure(figsize = (20,5))
sns.distplot(df['ConvertedComp'],kde = False)
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
plt.figure(figsize=(10,8))
ax = sns.boxplot(df['Age'],orient = 'v')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

plt.figure(figsize=(20,6))
sns.scatterplot(data=df,x='Age',y='WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

plt.figure(figsize=(20,6))
sns.scatterplot(data=df,size ='Age',y = 'CodeRevHrs',x = 'WorkWeekHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT *,count(1) as Count
FROM DatabaseDesireNextYear

group by DatabaseDesireNextYear
order by Count desc
limit 5
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
df_new = pd.read_sql_query(QUERY,conn)
df_new
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
# dfpie

df_new['Count'].plot(kind='pie',
                          figsize = (15,8),
                          autopct='%1.1f%%',
                          pctdistance= 1.13,
                          labels = None,
                          colors = colors_list,
                          shadow= True,
                          startangle =120)

plt.title('Top 5 databases that respondents wish to learn next year',y=1.12)
plt.axis('equal')
plt.legend(labels=df_new['DatabaseDesireNextYear'], loc='upper right') 
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

Query="""
    Select WorkWeekHrs, CodeRevHrs, Age
    from master 
    where Age between 30 and 35    
    """
dfage = pd.read_sql_query(Query,conn)
dfage = dfage.groupby('Age').median()
dfage
```

```python
dfage.plot(kind='bar',figsize=(15,6),stacked = True)

plt.xlabel('Age') # add to x-label to the plot
plt.ylabel('Number of hours') # add y-label to the plot
plt.title('Stacked Bar plot of Work week Hours and CodeRevHrs') # add title to the plot

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
        Select ConvertedComp,Age 
        from master
        where age between 45 and 60 """

dfcomp = pd.read_sql_query(QUERY,conn)
dfcomp = dfcomp.groupby('Age').median()
dfcomp
```

```python
dfcomp.plot(kind='line',figsize = (18,8))

plt.ylabel('Converted Comp')
plt.title('Plot of median ConvertedComp for all ages from 45 to 60')
plt.legend(loc = 'lower right')
plt.show()
```

Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """
        Select MainBranch,count(1) as Branch_Count
        from master
        group by MainBranch 
        """

df = pd.read_sql_query(QUERY,conn)
df.set_index('MainBranch')

df
```

```python

df.plot(kind='barh',figsize = (20,5))
plt.ylabel("Main Branch")
plt.show()
```
